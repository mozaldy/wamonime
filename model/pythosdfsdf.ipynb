```python
import re
import string

def separate(text):
    clean_text = []
    for t in text.split(','):
        cleaned = re.sub('\(.*\)', '', t) # Remove text inside parentheses
        cleaned = cleaned.translate(str.maketrans('','', string.digits))
        cleaned = cleaned.replace(' ', '')
        cleaned = cleaned.translate(str.maketrans('','', string.punctuation)).lower()
        clean_text.append(cleaned)
    return ' '.join(clean_text)

def remove_punc(text):
    cleaned = text.translate(str.maketrans('','', string.punctuation)).lower()
    clean_text = cleaned.translate(str.maketrans('','', string.digits))
    return clean_text
```

```python
anime['Genres'] = anime['Genres'].apply(remove_punc)
anime['Status'] = anime['Status'].apply(remove_punc)
anime['Producers'] = anime['Producers'].apply(separate)
anime['Licensors'] = anime['Licensors'].apply(remove_punc)
anime['Studios'] = anime['Studios'].apply(separate)
anime['Rating'] = anime['Rating'].apply(separate)

anime['bag_of_words'] = ''
anime['bag_of_words'] = anime[anime.columns[1:]].apply(lambda x: ' '.join(x), axis=1)
anime.set_index('original_title', inplace=True)

anime = anime[['bag_of_words']]
anime.head()
```
